#### Strip beginning and end of the text

````python
#export
def strip_text(df, txt_col="text"): df[txt_col] = df[txt_col].apply(lambda s: s.strip())
````

````python
strip_text(train_df, "text")
strip_text(train_df, "selected_text")
strip_text(test_df, "text")
````

````python
for s1,s2 in zip(train_df.text.values[:100], train_df.selected_text.values[:100]): print(s1+"|"+s2)
````

#### Remove trailing extra spaces

````python
#export
whitespace_pattern = re.compile(r"\s+")
def replace_whitespace(df, txt_col="text"): 
    df[txt_col] = df[txt_col].apply(lambda s: whitespace_pattern.sub(" ", s).strip())
````

````python
replace_whitespace(train_df, "text")
replace_whitespace(train_df, "selected_text")
replace_whitespace(test_df, "text")
````

````python
for s1,s2 in zip(train_df.text.values[:100], train_df.selected_text.values[:100]): print(s1+"|"+s2)
````

#### Replace url with "URL" token

````python
#export
url_pattern = re.compile("""(?:(?:https?|ftp):\/\/|\b(?:[a-z\d]+\.))(?:(?:[^\s()<>]+|\((?:[^\s()<>]+|(?:\([^\s()<>]+\)))?\))+(?:\((?:[^\s()<>]+|(?:\(?:[^\s()<>]+\)))?\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))?""")
def replace_URLs(df, txt_col="text", repl="URL"): 
    df[txt_col] = df[txt_col].apply(lambda s: url_pattern.sub(repl, s))
````

````python
replace_URLs(train_df, "text")
replace_URLs(train_df, "selected_text")
replace_URLs(test_df, "text")
````

````python
for s1,s2 in zip(train_df.text.values[:100], train_df.selected_text.values[:100]): print(s1+"|"+s2)
````

#### Replace user tags with "USER" token

````python
#export
user_pattern = re.compile("(?<!\w)[@_]\w+")
def replace_user(df, txt_col="text", repl="USER"): 
    df[txt_col] = df[txt_col].apply(lambda s: user_pattern.sub(repl, s))
````

````python
replace_user(train_df, "text")
replace_user(train_df, "selected_text")
replace_user(test_df, "text")
````

````python
for s1,s2 in zip(train_df.text.values[:500], train_df.selected_text.values[:500]): print(s1+"|"+s2)
````

### Drop data when there is partial selection of USER or URL

`get_start_end_idxs` fails in the case of:
        
```
context:'USER hi! **** your job!'
answer: 'and_jay hi! **** your job
```

````python
#export
def is_wrong_selection(context, answer):
    "Get string start and end char for answer span"
    len_a = len(answer)
    for i, _ in enumerate(context):
        if context[i:i+len_a] == answer: 
            return False
    return True
````

````python
is_wrong = train_df.apply(lambda o: is_wrong_selection(o['text'], o['selected_text']), 1)
````

````python
train_df[is_wrong]
````

````python
train_df = train_df[~is_wrong].reset_index(drop=True)
````
